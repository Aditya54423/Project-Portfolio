```python
#Rename column
data.rename(columns={'serum cholestoral':'serum cholesterol'},inplace=True)
data.head
```

```python
# Plotting distribution of 'age'
plt.figure(figsize=(10,6))
sns.histplot(data['age'], kde=True)
plt.title('Distribution of Age (Heart Disease)')
plt.xlabel('Age')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
# Plotting distribution of 'chol' (serum cholestoral)
plt.figure(figsize=(10,6))
sns.histplot(data['serum cholesterol'], kde=True)
plt.title('Distribution of Serum Cholesterol')
plt.xlabel('Cholesterol')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix
plt.figure(figsize=(14,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Heart Disease Dataset')
plt.tight_layout()
plt.show()
```

```python
# Pair plot for relationships between features
sns.pairplot(data, hue='age', diag_kind='kde')
plt.suptitle('Pair Plot of Heart Disease Dataset', y=1.02,fontsize = 25)
plt.show()
# Scatter plot for age vs. max heart rate
plt.figure(figsize=(10,6))
sns.scatterplot(x='age', y='thal', hue='age', data=data)
plt.title('Age vs. Max Heart Rate (Thalach)')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate')
plt.tight_layout()
plt.show()
```
